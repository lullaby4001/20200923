hanoi (漢諾塔)

```
d=1
def hanoi(n ,A,B,C):
  global d
  
  if n == 1 :
    print("%c-->%c 執行次數:%d" %(A,C,d))
    d=d+1
  else:
    hanoi(n-1,A,C,B)
    hanoi(1,A,B,C)
    hanoi(n-1,B,A,C)
j=int(input("輸入套環數量:"))
hanoi(j,"A","B","C")
```

有問題的漢諾塔

```
def hannoi(n,x,y,z):
    if n==1:  #設置何時終止「將N移動到結束區」的重複（即漢諾塔的層數只有一層的時候，直接將漢諾塔從起始區移動到終止區即可）。因此，n = 1 時，print (x ,'-->', z)
        print(x,'-->',z)
    else:
        hannoi(n-1, x, z, y)	#將第n層以上的層次看做一個整體N，從起始區移動到中轉區
        print(x,'-->',z)		#將第n層移動到結束區
        for i in range(n):
          print('執行次數：',i+1)
          hannoi(n-1, y, x, z)	#將N移動到結束區（即：將起始區和中轉區互換，重複以上步驟）

n=int(input("請輸入漢諾塔的層級數"))   
hannoi(n, "X", "Y", "Z")
```
